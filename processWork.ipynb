```python
class my_dictionary(dict): 
  
    # __init__ function 
    def __init__(self): 
        self = dict() 
          
    # Function to add key:value 
    def add(self, key, value): 
        self[key] = value
        
    def checkKey(self, key): 
      
        if key in self.keys(): 
            return True 
        else: 
            return False
        
    # Increment value
    def increment(self,key):
        
        if key in self.keys():
            
            self[key] += 1
    
```
